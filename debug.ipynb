```python
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
```

```python
df = pd.read_excel('city_price.xlsx')
df["CITY"] = list(map(lambda city: city.split(",")[-1], df["ADDRESS"]))
```

```python
df
```

```python
dummies = pd.get_dummies(df.CITY)
```

```python
dummies_df = pd.concat([df, dummies], axis='columns')
```

```python
dummies_df
```

```python
X = dummies_df.drop(columns=["POSTED_BY", "BHK_OR_RK", "RERA", "RESALE", "ADDRESS", "LONGITUDE", "LATITUDE", "TARGET(PRICE_IN_LACS)", "CITY", "Bangalore"])
Y = df["TARGET(PRICE_IN_LACS)"]
```

```python
X
```

```python
knr_model = KNeighborsRegressor()
knr_model.fit(X, Y)
```

```python
knr_model.predict([[0, 2, 1300.236407, 1, 0, 0, 1, 0, 0, 0, 0]])
```

```python
knr_model.score(X, Y)
```
